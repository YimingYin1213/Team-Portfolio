# Basketball Game Breakdown

## 1. Level Setup

The class starts by storing `gameEnv`, screen size, image paths, player spawn points, timing values, projectile settings, and UI references.

A big part of the constructor is building `this.classes`, which tells the engine what objects should exist in the level.

```javascript
this.classes = [
  { class: GameEnvBackground, data: image_data_court },
  { class: Player, data: sprite_data_player },
  { class: Npc, data: sprite_data_chaser },
  { class: Coin, data: coin_1 },
  { class: Coin, data: coin_2 },
  { class: Coin, data: coin_3 },
  { class: Barrier, data: barrier_bench_top },
  { class: Barrier, data: barrier_bench_bottom },
  { class: Barrier, data: barrier_gatorade_left },
  { class: Barrier, data: barrier_gatorade_right }
];
```

This array is important because it is the engine's blueprint for the basketball court. It creates the background, the player, the chaser, the coins, and the invisible collision zones.


## 2. Game Flow and Control Structures

The `initialize()` method prepares the level state, creates the HUD, shows the intro dialogue, and attaches keyboard listeners.

The `update()` method acts like the main game loop. Each frame it:

- finds the player and LeBron objects
- updates the projectile positions
- stops motion if the intro screen is still active
- tracks survival time
- moves LeBron toward the player
- checks whether the player has been caught

```javascript
const dx = player.position.x - lebron.position.x;
const dy = player.position.y - lebron.position.y;
const dist = Math.hypot(dx, dy);

const speed = Math.min(2.1 + this.currentTime * 0.03, 2.8);
lebron.position.x += (dx / dist) * speed;
lebron.position.y += (dy / dist) * speed;
```

That logic uses variables, arithmetic, conditionals, and a little geometry so the chaser follows the player in a fair way instead of teleporting.


## 3. Input, Projectiles, and Collision Logic

The player can press `E` to shoot a basketball. The method `handleShootKey()` checks the key pressed, makes sure the player is allowed to shoot, and respects the cooldown.

Then `spawnProjectileFromPlayer()` creates a projectile object with position, direction, speed, lifetime, and its own canvas element.

```javascript
const projectile = {
  x: player.position.x + (player.width || 0) / 2,
  y: player.position.y + (player.height || 0) / 2,
  vx: directionVector.x * this.projectileSpeed,
  vy: directionVector.y * this.projectileSpeed,
  radius: this.projectileRadius,
  bornAt: now,
  canvas: document.createElement('canvas')
};
```

Later, `updateProjectiles()` loops backward through the array of projectiles, moves each one, removes expired shots, and checks if a ball hit LeBron.

The file also has helper methods like `getHitboxRect()`, `isHitboxCollision()`, and `isCircleHittingObject()` so collisions are more accurate and easier to reuse.


## 4. UI, Scores, and Reset Logic

The game is not only sprites moving on a court. This file also creates a full gameplay experience with UI and persistence.

- `createHud()` builds the timer and message boxes
- `updateHud()` refreshes time, best time, coins, and best coin total
- `initLeaderboard()` connects the level to the leaderboard system
- `submitRoundScore()` sends the score after the player is caught
- `resetRound()` restores player position, enemy position, coin locations, timer values, and projectile state
- `loadBestTime()` and `saveBestTime()` use `localStorage` so best results stay saved

This is a good example of decomposition: instead of putting everything in one huge method, the file separates HUD work, movement logic, collision checks, score saving, and cleanup into focused functions.
